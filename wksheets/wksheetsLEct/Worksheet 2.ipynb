```python
#1.	Colour splitting
#Write a program which reads/loads an RGB image and displays the red, green, and blue channels separately, each one of them as an RGB image.
from PIL import Image
import matplotlib.pyplot as plt #to display image
import numpy as np

im = np.array(Image.open('lena.jpg'))
plt.imshow(im)
plt.show()

im_R = im.copy()
im_R[:, :, (1, 2)] = 0
im_G = im.copy()
im_G[:, :, (0, 2)] = 0
im_B = im.copy()
im_B[:, :, (0, 1)] = 0

plt.imshow(im_R)
plt.show()

plt.imshow(im_G)
plt.show()

plt.imshow(im_B)
plt.show()
```

```python
#2.	Image Brightening
#   Write a program which loads an RGB image and brightens each pixel by a
#   value of 5 (current brightness + 5).
#   Do this by applying an addition operation between a matrix and a scalar.

from PIL import Image
import matplotlib.pyplot as plt #to display image
import numpy as np

im = np.array(Image.open('lena.jpg'))

im += 50

plt.imshow(im)
plt.show()
```

```python
#3.	Image trimming
#Write a program which loads an RGB image and trims out a margin of 20 pixels
# from all around the image. Do this by using the colon notation.
# Display the new image.

from PIL import Image
import matplotlib.pyplot as plt #to display image
import numpy as np

im = np.array(Image.open('lena.jpg'))

#with buffer .. same image size
im[:20,:,:]= 0
im[-20:,:,:]= 0
im[:,:20,:]= 0
im[:,-20:,:]= 0

plt.figure()
plt.imshow(im)
plt.show()

#cropping 
im= im[20:-20,20:-20,:]
plt.figure()
plt.imshow(im)
plt.show()
```

```python
#4.	Image flipping
#Write a program which loads an RGB image and flips it both vertically and horizontally. Do this by using the colon notation. Display the new image.
import numpy as np
from PIL import Image

im = np.array(Image.open('lena.jpg'))

#Display bottom up flip example


upflipped = im[::-1, :, :]
plt.imshow(upflipped)
plt.show()

leftflipped = im[:,::-1, :]
plt.imshow(leftflipped)
plt.show()

```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg')

imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)
plt.title('Bright')
plt.show()


imghsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

h , s , v = cv2.split(imghsv)

value = 50

limit = 255 - value
v[v>limit] = 255
v[v<=limit] += value

bright_hsv = cv2.merge((h,s,v))
imgbrightRGB = cv2.cvtColor(bright_hsv, cv2.COLOR_HSV2RGB)

plt.imshow(imgbrightRGB)
plt.title('Bright')
plt.show()
```
