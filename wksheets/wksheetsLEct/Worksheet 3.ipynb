```python
#Task 2.	Colour splitting with subplots
#Write a program which reads/loads an RGB image and displays the red, green, and blue channels separately, each one of them as an RGB image.
from PIL import Image
import matplotlib.pyplot as plt #to display image
import numpy as np

im = np.array(Image.open('lena.jpg'))
plt.figure(figsize= (10,10))
plt.subplot(221)
plt.axis('off')
plt.imshow(im)


im_R = im.copy()
im_R[:, :, (1, 2)] = 0

im_G = im.copy()
im_G[:, :, (0, 2)] = 0
im_B = im.copy()
im_B[:, :, (0, 1)] = 0

#mag = 255 - im_G;

plt.subplot(222)
plt.axis('off')
plt.imshow(im_R)

plt.subplot(223)
plt.axis('off')
plt.imshow(im_G)

plt.subplot(224)
plt.axis('off')
plt.imshow(im_B)

plt.show()

Image.fromarray(im_R).save('lena_RED.jpg')
Image.fromarray(im_G).save('lena_GREEN.jpg')
Image.fromarray(im_B).save('lena_BLUE.jpg')
```

```python
#Task 3 - Splitting

from PIL import Image
import matplotlib.pyplot as plt #to display image
import numpy as np

im = np.array(Image.open('lena.jpg'))

plt.imshow(im)
plt.show()

height, width, channels = np.shape(im)

plt.figure(figsize= (10,10))

im_TL = im[:(height // 2), :(width // 2),:]

plt.subplot(2,2,1)
plt.axis('off')
plt.imshow(im_TL)

im_TR = im[:(height // 2), (width // 2):,:]
plt.subplot(2,2,2)
plt.axis('off')
plt.imshow(im_TR)

im_BL = im[(height // 2):, :(width // 2),:]
plt.subplot(2,2,3)
plt.axis('off')
plt.imshow(im_BL)

im_BR = im[(height // 2):, (width // 2):,:]
plt.subplot(2,2,4)
plt.axis('off')
plt.imshow(im_BR)

plt.show()
```
