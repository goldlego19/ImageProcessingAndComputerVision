```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
#task1

imBGR = cv2.imread('../../images/lena.jpg')
imGRAY = cv2.cvtColor(imBGR,cv2.COLOR_BGRA2GRAY) 


bins = 256
flattened_imGray = imGRAY.ravel()  # change the matrix to a single dimensional list called flattening the image
plt.subplot(1,2,1)
plt.title("Grayscale image")
plt.imshow(imGRAY, cmap="gray", vmin=0, vmax=255)


plt.subplot(1,2,2)

plt.hist(flattened_imGray, bins, range=[0,255])
plt.xlabel("Pixel Intensity")
plt.ylabel("No of pixels")
plt.title("Image Histogram")
plt.show()
```

```python
#Task 2

imBGR = cv2.imread('dots.bmp')
imGRAY = cv2.cvtColor(imBGR,cv2.COLOR_BGRA2GRAY) 
imRGB = cv2.cvtColor(imBGR,cv2.COLOR_BGR2RGB)

plt.imshow(imGRAY, cmap="gray", vmin=0,vmax=255)

rows, cols = imGRAY.shape

count=0

for x in range(rows):
    for y in range(cols):
        if (imGRAY[x,y]==0) :
            count += 1
            cv2.circle(imRGB, center=(y,x), radius=10, color=(255,0,0), thickness=3)

label = 'No of pixels: ' + str(count)
cv2.putText(imRGB, label, (10,30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2)

plt.imshow(imRGB)
```

```python
rows, cols = 15, 15
image = np.random.choice([0,255], (rows,cols,3)).astype(np.uint8)

image_resized = cv2.resize(image, (150,150), interpolation=cv2.INTER_NEAREST)

cell_size = 10

red_count = green_count = blue_count = 0

for cell_y in range(15):
    for cell_x in range(15):

        # compute pixel coordinates of the center
        px = cell_x * cell_size + cell_size // 2
        py = cell_y * cell_size + cell_size // 2

        r, g, b = image_resized[py, px]

        if (r, g, b) == (255, 0, 0):
            cv2.circle(image_resized, (px, py), 8, (255,0,0), 1)
            red_count += 1

        elif (r, g, b) == (0, 255, 0):
            cv2.circle(image_resized, (px, py), 8, (0,255,0), 1)
            green_count += 1

        elif (r, g, b) == (0, 0, 255):
            cv2.circle(image_resized, (px, py), 8, (0,0,255), 1)
            blue_count += 1


plt.imshow(image_resized)
plt.show()

print(red_count)
print(green_count)
print(blue_count)
```
