```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
possible_values = np.array([50,140,190], dtype='uint8')
 
randomimg = np.random.choice(possible_values, size=[10,10])
print(randomimg)
 
plt.subplot(1,2,1)
plt.title('Original image')
plt.imshow(randomimg, cmap='gray', vmin=0, vmax=255)
 
plt.subplot(1,2,2)
plt.title('Original histogram')
hist=cv2.calcHist([randomimg], [0], None, [256], [0,256])
plt.plot(hist)
 
plt.show()
 
 
normalized = cv2.normalize(randomimg, None, 0, 255, cv2.NORM_MINMAX)
plt.subplot(1,2,1)
plt.title('Normalized image')
plt.imshow(normalized, cmap='gray', vmin=0, vmax=255)
 
plt.subplot(1,2,2)
plt.title('Original histogram')
hist=cv2.calcHist([normalized], [0], None, [256], [0,256])
plt.plot(hist)
 
plt.show()
```
