Task 1 - Thresholding


```python
import cv2
import imutils
import matplotlib.pyplot as plt
import numpy as np


# load the image from disk
img = cv2.imread('../../images/trees.png',0)

thresh1 = 25 * 255 // 100
thresh2 = 50 * 255 // 100
thresh3 = 75 * 255 // 100

ret,img_t1 = cv2.threshold(img ,thresh1, 255,cv2.THRESH_BINARY)
ret,img_t2 = cv2.threshold(img ,thresh2, 255,cv2.THRESH_BINARY)
ret,img_t3 = cv2.threshold(img ,thresh3, 255,cv2.THRESH_BINARY)

plt.figure(figsize=(15,15))
titles=['Original', '25%', '50%', '75%']
images = [img, img_t1, img_t2, img_t3]

for count in range(4):
    plt.subplot(2,2,count+1)
    plt.imshow(images[count],cmap='gray')
    plt.title(titles[count])
    plt.axis('off')

plt.show()
```

Task 2 - Adaptive thresholding

```python
import cv2
import imutils
import matplotlib.pyplot as plt
import numpy as np


# load the image from disk
img = cv2.imread('creative-proposal.jpg',0)

thresh = 50 * 255 // 100

ret,img_t1 = cv2.threshold(img ,thresh, 255,cv2.THRESH_BINARY)
img_t2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)
img_t3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)

titles=['Original', '50%', 'Adaptive Mean', 'Adaptive Gaussian']
images = [img, img_t1, img_t2, img_t3]
plt.figure(figsize=(15,15))

for count in range(4):
    plt.subplot(2,2,count+1)
    plt.imshow(images[count],cmap='gray')
    plt.title(titles[count])
    plt.axis('off')

plt.show()
```

Task 3 - Morphology

```python
import cv2
import imutils
import matplotlib.pyplot as plt
import numpy as np

# load the image from disk as grayscale
image = cv2.imread('j.png', 0)
ret,image= cv2.threshold(image ,127, 255,cv2.THRESH_BINARY)

kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(image,kernel,iterations = 1)

dilation = cv2.dilate(image,kernel,iterations = 1)

plt.figure(figsize=(15,15))
titles=['Original', 'Erosion', 'Dilation']
images = [image, erosion, dilation]

for count in range(3):
    plt.subplot(1,3,count+1)
    plt.imshow(images[count],cmap='gray')
    plt.title(titles[count])
    plt.axis('off')

plt.show()
```

Task 4 - Opening and closing

```python
import cv2
import imutils
import matplotlib.pyplot as plt
import numpy as np

# load the image from disk as grayscale
image1 = cv2.imread('opening_original.png', 0)
ret,image1= cv2.threshold(image1 ,127, 255,cv2.THRESH_BINARY)

kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(image1, cv2.MORPH_OPEN, kernel)

# load the image from disk as grayscale
image2 = cv2.imread('closing_original.png', 0)
ret,image2= cv2.threshold(image2 ,127, 255,cv2.THRESH_BINARY)
closing = cv2.morphologyEx(image2, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(20,20))
titles=['Original', 'Opening', 'Original', 'Closing']
images = [image1, opening, image2, closing]

for count in range(4):
    plt.subplot(2,2,count+1)
    plt.imshow(images[count],cmap='gray')
    plt.title(titles[count])
    plt.axis('off')

plt.show()
```

Task 5 - Masking Example

```python
import cv2
import imutils
import matplotlib.pyplot as plt
import numpy as np


src1 = cv2.imread('images/lena.jpg')
src2 = cv2.imread('images/horse_r.png')
src1 = cv2.cvtColor(src1, cv2.COLOR_BGR2RGB)
src2 = cv2.cvtColor(src2, cv2.COLOR_BGR2RGB)

src2 = cv2.resize(src2, src1.shape[1::-1])

print(src2.shape)
# (225, 400, 3)

print(src2.dtype)
# uint8

dst = cv2.bitwise_and(src1, src2)

plt.figure(figsize=(15,15))
titles=['Original', 'Mask', 'Result']
images = [src1, src2, dst]

for count in range(3):
    plt.subplot(1,3,count+1)
    plt.imshow(images[count])
    plt.title(titles[count])
    plt.axis('off')


dst = cv2.cvtColor(dst, cv2.COLOR_RGB2BGR)
cv2.imwrite('opencv_bitwise_and.jpg', dst)

```
