```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
##Colour Splitting
imBGR = cv2.imread("../../images/lena.jpg") # as Default the image is read as BGR instead of RGB
imRGB = cv2.cvtColor(imBGR,cv2.COLOR_BGR2RGB) # this converts from BGR to RGB

redchannel= imRGB.copy()
redchannel[:,:,1]=0
redchannel[:,:,2]=0

greenchannel= imRGB.copy()
greenchannel[:,:,0]=0
greenchannel[:,:,2]=0

bluechannel= imRGB.copy()
bluechannel[:,:,0]=0
bluechannel[:,:,1]=0

plt.imshow(redchannel) 
plt.title("Red Channel")
plt.show()

plt.imshow(greenchannel)
plt.title("Green Channel")
plt.show()

plt.imshow(bluechannel)
plt.title("Blue Channel")
plt.show()
```

```python
#image brightening

imBGR = cv2.imread("../../images/lena.jpg")
imRGB = cv2.cvtColor(imBGR,cv2.COLOR_BGR2RGB)

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(imRGB)

brightness=5
imageBright = np.clip(imRGB.astype(int)+brightness,0,255)

plt.subplot(1, 2, 2)
plt.title("Brightness ")
plt.imshow(imageBright)
plt.show()


```

```python
#image trimming

imBGR = cv2.imread("../../images/lena.jpg")
imRGB = cv2.cvtColor(imBGR,cv2.COLOR_BGR2RGB)

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(imRGB)

imageTrimmed = imRGB[20:-20,20:-20,:]

plt.subplot(1, 2, 2)
plt.title("Trimmed ")
plt.imshow(imageTrimmed)
plt.show()
```

```python
#image flipping

imBGR = cv2.imread("../../images/lena.jpg")
imRGB = cv2.cvtColor(imBGR,cv2.COLOR_BGR2RGB)

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(imRGB)

imageFlipped = imRGB[::-1,::-1, :]

plt.subplot(1, 2, 2)
plt.title("Flipped")
plt.imshow(imageFlipped)
plt.show()
```
