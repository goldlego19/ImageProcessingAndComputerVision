```python
#imports
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
m1 = np.array([[2,4,6,8], [10,20,30,40],[5,10,15,20]])

print("2nd row of M by using the colon operator")
print(m1[1,:])

print("\n4th column of M by using the colon operator")
print(m1[:,3])
print("\nAssign all elements of M greater than 9 the value 0")
m1[m1>9] = 0
print(m1)

```

```python

print(m1.shape)
m2 = np.random.randint(0, 256, size=m1.shape, dtype='uint8')

plt.imshow(m2, cmap='gray', vmin=0, vmax=255)

```

```python
#q6
m3 = np.zeros([100,200,3], dtype='uint8')

print(m3.shape)
print(f'Matrix type: \n{m3.dtype}')

m3[:,:100] =[255,128,0]
m3[:,100:] =[0,0,255]


print(m3)

print('My RGB matrix as an image')
plt.imshow(m3)
plt.show()

save_output = Image.fromarray(m3)
save_output.save('Q6.bmp')
```

```python
#q7
m4 = np.zeros([100,200,4], dtype='uint8')

print(m4.shape)
print(f'Matrix type: \n{m4.dtype}')

m4[:,:100] =[255,128,0,0]
m4[:,100:] =[0,0,255,255]

i=np.linspace(0,255,200,dtype='uint8')

for x in range(200):
    m4[:,x,3] = i[x]

print('My RGB matrix as an image')
plt.imshow(m4)
plt.show()

#0 transparency means fully transparent whilst the other extreme is 255 is fully visible 

```
