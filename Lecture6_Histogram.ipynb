```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Perform histogram normalization and display the output image and histogram side-by-side in a 1 by 2 subplot
possible_values = np.array([50,140,160,180,190], dtype='uint8')
 
randomimg = np.random.choice(possible_values, size=[10,10])
print(randomimg)
 
plt.subplot(1,2,1)
plt.title('Original image')
plt.imshow(randomimg, cmap='gray', vmin=0, vmax=255)
 
plt.subplot(1,2,2)
plt.title('Original histogram')
hist=cv2.calcHist([randomimg], [0], None, [256], [0,256])
plt.plot(hist)
 
plt.show()
 
 
normalized = cv2.normalize(randomimg, None, 0, 255, cv2.NORM_MINMAX)
plt.subplot(1,2,1)
plt.title('Normalized image')
plt.imshow(normalized, cmap='gray', vmin=0, vmax=255)
 
plt.subplot(1,2,2)
plt.title('Normalized histogram')
hist=cv2.calcHist([normalized], [0], None, [256], [0,256])
plt.plot(hist)
 
plt.show()
```

```python
#Perform histogram equalization and display the output image and histogram side-by-side in a 1 by 2 subplot
possible_values = np.array([50,140,160,180,190], dtype='uint8')
 
randomimg = np.random.choice(possible_values, size=[10,10])
print(randomimg)
 
plt.subplot(1,2,1)
plt.title('Original image')
plt.imshow(randomimg, cmap='gray', vmin=0, vmax=255)
 
plt.subplot(1,2,2)
plt.title('Original histogram')
hist=cv2.calcHist([randomimg], [0], None, [256], [0,256])
plt.plot(hist)
 
plt.show()
 
 
equalized = cv2.equalizeHist(randomimg)
plt.subplot(1,2,1)
plt.title('Equalized image')
plt.imshow(equalized, cmap='gray', vmin=0, vmax=255)
 
plt.subplot(1,2,2)
plt.title('Equalized histogram')
hist=cv2.calcHist([equalized], [0], None, [256], [0,256])
plt.plot(hist)
 
plt.show()
```

```python
#Normalizeing the brightness channel using HSV color space.


imBGR = cv2.imread('images/locontrast.jpg')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
imHSV = cv2.cvtColor(imBGR, cv2.COLOR_BGR2HSV)

plt.subplot(1,2,1)
plt.title('Original image')
plt.imshow(imRGB)

 
plt.subplot(1,2,2)
plt.title('Original histogram')
hist = cv2.calcHist([imRGB],[0], None, [256],[0,256]) 
plt.plot(hist, color=[1,0,0])
hist = cv2.calcHist([imRGB],[1], None, [256],[0,256]) 
plt.plot(hist, color=[0,1,0])
hist = cv2.calcHist([imRGB],[2], None, [256],[0,256]) 
plt.plot(hist, color=[0,0,1])
 
plt.show()
 
 
imHSV[:,:,2] = cv2.normalize(imHSV[:,:,2], None, 0, 255, cv2.NORM_MINMAX)
outRGB = cv2.cvtColor(imHSV, cv2.COLOR_HSV2RGB)
plt.subplot(1,2,1)
plt.title('Normalized image')
plt.imshow(outRGB)
 
plt.subplot(1,2,2)
plt.title('Normalized histogram')
hist = cv2.calcHist([outRGB],[0], None, [256],[0,256]) 
plt.plot(hist, color=[1,0,0])
hist = cv2.calcHist([outRGB],[1], None, [256],[0,256]) 
plt.plot(hist, color=[0,1,0])
hist = cv2.calcHist([outRGB],[2], None, [256],[0,256]) 
plt.plot(hist, color=[0,0,1])
 
plt.show()
```
