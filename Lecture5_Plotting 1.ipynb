```python
imBGR = cv2.imread('images/locontrast.jpg')
imGray = cv2.cvtColor(imBGR, cv2.COLOR_BGR2GRAY)
plt.imshow(imGray, cmap='grey', vmin=0, vmax=255)
plt.show()

bins = 256
flattened_imGray = imGray.ravel()  # change the matrix to a single dimensional list called flattening the image
print(flattened_imGray)

plt.hist(flattened_imGray, bins, range=[0,255])
plt.xlabel("Pixel Intensity")
plt.ylabel("No of pixels")
plt.title("Image Histogram")
plt.show()
```

```python
#Histogram drawing using OPENCV

imBGR = cv2.imread('images/locontrast.jpg')
imGray = cv2.cvtColor(imBGR, cv2.COLOR_BGR2GRAY)
plt.imshow(imGray, cmap='grey', vmin=0, vmax=255)
plt.show()

hist = cv2.calcHist([imGray],[0], None, [256],[0,256]) #calcHist expects a list of images. Do not forget []
plt.plot(hist)
plt.show()
```

```python
#Color histogram drawing with OPENCV

imBGR = cv2.imread('images/goldfish.jpg')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
plt.imshow(imRGB)
plt.show()

hist = cv2.calcHist([imRGB],[0], None, [256],[0,256]) 
plt.plot(hist, color=[1,0,0])
hist = cv2.calcHist([imRGB],[1], None, [256],[0,256]) 
plt.plot(hist, color=[0,1,0])
hist = cv2.calcHist([imRGB],[2], None, [256],[0,256]) 
plt.plot(hist, color=[0,0,1])

plt.show()

```

```python
imBGR = cv2.imread('images/monkey.jpg')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
#plt.imshow(imRGB)
#plt.show()

out_RGB = imRGB.copy()
x, y = 270,280
cv2.circle(out_RGB, center=(x,y), radius=200, color=(255,0,0), thickness= 3)


font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(out_RGB,'I am your lecturer',(10,30), font, 1,(0,0,255),2,cv2.LINE_AA)

plt.imshow(out_RGB)
plt.show()
```
