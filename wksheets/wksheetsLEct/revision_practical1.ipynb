```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import cv2

gradient = np.zeros ((100,256),dtype=np.uint8)

for i in range(0,256):
    gradient[:,i] = i;

print(gradient)
plt.imshow(gradient,cmap=cm.gray)
plt.show()

hist = cv2.calcHist([gradient],[0],None,[256],[0,255])
plt.plot(hist)
plt.title("Histogram of gradient image")
plt.show()
```

```python
# TASK 3 (DOWNSAMPLING)
import numpy as np

bgr_img = cv2.imread('images\lena.jpg')

bgr_img = bgr_img[::2,::2]

plt.title("Down sampled by half")
rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
```

```python
# TASK 4 (UPSAMPLING)
import numpy as np

bgr_img = cv2.imread('images\lena.jpg')
rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)

original_height = rgb_img.shape[0]
new_height = rgb_img.shape[0]*2

original_width = rgb_img.shape[1]
new_width = rgb_img.shape[1]*2

longer = np.zeros((new_height,original_width,3), np.uint8)

for x in range(original_height):
    longer[x*2] = rgb_img[x]
    longer[(x*2)+1] = rgb_img[x]

longerandwider = np.zeros((new_height,new_width,3), np.uint8)

for y in range(original_width):
    longerandwider[:,y*2] = longer[:,y]
    longerandwider[:,(y*2)+1] = longer[:,y]

plt.title("Up sampled by double")

plt.imshow(longerandwider)
plt.show()
```
