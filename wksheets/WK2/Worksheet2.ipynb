```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
##Colour Splitting
imBGR = cv2.imread("../../images/lena.jpg") # as Default the image is read as BGR instead of RGB
imRGB = cv2.cvtColor(imBGR,cv2.COLOR_BGR2RGB) # this converts from BGR to RGB
plt.imshow(imRGB[:,:,0], cmap='Reds') 
plt.title("Red Channel")
plt.show()

plt.imshow(imRGB[:,:,1], cmap='Greens')
plt.title("Green Channel")
plt.show()

plt.imshow(imRGB[:,:,2], cmap='Blues')
plt.title("Blue Channel")
plt.show()
```

```python
#image brightening

imBGR = cv2.imread("../../images/lena.jpg")
imRGB = cv2.cvtColor(imBGR,cv2.COLOR_BGR2RGB)

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(imRGB)

brightness = 30
contrast = 1  
imageBright = cv2.addWeighted(imRGB, contrast , np.zeros(imRGB.shape, imRGB.dtype),0, brightness)

plt.subplot(1, 2, 2)
plt.title("Brightness ")
plt.imshow(imageBright)
plt.show()


```

```python
#image trimming

imBGR = cv2.imread("../../images/lena.jpg")
imRGB = cv2.cvtColor(imBGR,cv2.COLOR_BGR2RGB)

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(imRGB)

imageTrimmed = imRGB[20:-20,20:-20,:]

plt.subplot(1, 2, 2)
plt.title("Trimmed ")
plt.imshow(imageTrimmed)
plt.show()
```

```python
#image flipping

imBGR = cv2.imread("../../images/lena.jpg")
imRGB = cv2.cvtColor(imBGR,cv2.COLOR_BGR2RGB)

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(imRGB)

imageFlipped = imRGB[::-1,::-1, :]

plt.subplot(1, 2, 2)
plt.title("Flipped")
plt.imshow(imageFlipped)
plt.show()
```
