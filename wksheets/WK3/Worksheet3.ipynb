```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python

#TASK1
imBGR = cv2.imread('../../images/lena.jpg') 
imGRAY = cv2.cvtColor(imBGR,cv2.COLOR_BGRA2GRAY) 
imGRAYNPARRAY = np.array(imGRAY,dtype=np.uint8)


t_value=int(input("Please input an integer"))

if (t_value>=0 and t_value<=255):
    invertedGray = np.clip(t_value - imGRAYNPARRAY.astype(int),0,255)
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(imGRAYNPARRAY, cmap='gray')

    plt.subplot(1, 2, 2)
    plt.title("inverted")
    plt.imshow(invertedGray, cmap='gray')

    plt.show()

print("Threshold need to be between 0 and 255")



```

```python
#TASK2
imBGR = cv2.imread('../../images/lena.jpg') 
imRGB = cv2.cvtColor(imBGR,cv2.COLOR_BGR2RGB)

redchannel= imRGB.copy()
redchannel[:,:,1]=0
redchannel[:,:,2]=0

greenchannel= imRGB.copy()
greenchannel[:,:,0]=0
greenchannel[:,:,2]=0

bluechannel= imRGB.copy()
bluechannel[:,:,0]=0
bluechannel[:,:,1]=0


plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(imRGB)

plt.subplot(1, 2, 2)
plt.title("Reds")
plt.imshow(redchannel)

plt.show()
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(imRGB)

plt.subplot(1, 2, 2)
plt.title("Greens")
plt.imshow(greenchannel)

plt.show()

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(imRGB)

plt.subplot(1, 2, 2)
plt.title("Blues")
plt.imshow(bluechannel)

plt.show()


plt.imsave("Task2Reds.jpg",redchannel,cmap="Reds",vmin=0,vmax=255)
plt.imsave("Task2Greens.jpg",greenchannel,cmap="Greens",vmin=0,vmax=255)
plt.imsave("Task2Blues.jpg",bluechannel,cmap="Blues",vmin=0,vmax=255)

```
