```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imBGR = cv2.imread('../../images/lines.jpg')
imGray = cv2.cvtColor(imBGR,cv2.COLOR_BGR2GRAY)
imRGB = cv2.cvtColor(imBGR,cv2.COLOR_BGR2RGB)



_,binary = cv2.threshold(imGray,150,255,cv2.THRESH_BINARY)
plt.imshow(binary,'gray',vmin=0,vmax=255)
plt.title('Binary Image using THRESH_BINARY')
plt.show()

contours, _ = cv2.findContours(binary,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

objects_count = 0
for c in contours:
    area = cv2.contourArea(c)
    print(f'Contour area: {area}')
    if area > 50:
        if area < 52000:
            objects_count += 1
            cv2.drawContours(imRGB, [c], -1, (0,255,0), 2)



cv2.putText(imRGB, f'Objects: {objects_count}', (10,30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2)

plt.imshow(imRGB)
plt.title('All Contours drawn')
plt.show()


```
