Task 1 - Plotting grayscale image histogram:


```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

myimage = cv2.imread('./images/lena.jpg')
fig,axs = plt.subplots(1,2)
grayimage = cv2.cvtColor(myimage, cv2.COLOR_BGR2GRAY)
axs[0].imshow(grayimage, cmap=plt.cm.gray)
axs[1].hist(grayimage.ravel(),256,[0,256]);
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

img = cv2.imread('./images/black_dots.bmp')
dots_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


height, width = dots_img.shape
print(height, width)
count = 0

for x in range(height):
    for y in range(width):
        if dots_img[x,y] == 0:
            count+=1
            cv2.circle(img, center=(y,x), radius=10, color=(255,0,0), thickness=3)
label = 'No of pixels: ' + str(count)
cv2.putText(img, label, (10,30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2)
plt.imshow(img)
plt.show()
print(count)

```
